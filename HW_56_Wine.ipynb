```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_wine
```

```python
data = load_wine()
type(data)
```

```python
data_keys = data.keys()
data_keys
```

```python
feature_names = data["feature_names"]
feature_names
```

```python
target_names = data["target_names"]
target_names
```

#### Целевая переменная содержит три класса: 'class_0', 'class_1', 'class_2'

```python
print(data["DESCR"])
```

```python
X = pd.DataFrame(data.data, columns=feature_names)
X.info()
```

```python
X.shape
```

```python
X['target'] = data.target.astype('int64')
```

```python
X.describe().T
```

```python
X_corr = X.corr()
X_corr
```

```python
high_corr = X_corr[(X_corr.abs() > 0.5)].target[:-1].sort_values(ascending=False)
high_corr
```

```python
type(high_corr)
```

```python
X.drop(["target"], axis=1, inplace=True)
```

```python
X.info()
```

```python
for i in high_corr.index:
        X[f'{i}_2'] = X[i]**2
```
